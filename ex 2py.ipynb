Q8

```python
import math

# Program to calculate area of a triangle using Heron's formula

def triangle_area(a, b, c):
    # Semi-perimeter
    s = (a + b + c) / 2
    
    # Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

# Example test data
a = 5
b = 6
c = 7

area = triangle_area(a, b, c)
print(f"Area of the triangle = {area}")
```

```python
# Program to convert seconds into hours, minutes, and seconds

def convert_seconds(total_seconds):
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    seconds = total_seconds % 60
    return hours, minutes, seconds

# Example test data
total_seconds = 3665  # 1 hour, 1 minute, 5 seconds
h, m, s = convert_seconds(total_seconds)

print(f"{total_seconds} seconds = {h} hours, {m} minutes, {s} seconds")
```

Q10

```python
# Program to swap two numbers without using an additional variable

def swap_numbers(a, b):
    print(f"Before swap: a = {a}, b = {b}")
    
    # Swapping using addition and subtraction
    a = a + b
    b = a - b
    a = a - b
    
    print(f"After swap: a = {a}, b = {b}")
    return a, b

# Example test data
a = 5
b = 10
swap_numbers(a, b)
```

```python
# Program to print truth table for bitwise operators &, |, ^

def truth_table():
    print("A B | A & B | A | B | A ^ B")
    print("-----------------------------")
    for a in [0, 1]:
        for b in [0, 1]:
            print(f"{a} {b} |   {a & b}   |   {a | b}   |   {a ^ b}")

# Run the truth table
truth_table()
```

Q13

```python
# Program to find left shift and right shift values of a given number

def shift_operations(num, shift_by):
    left_shift = num << shift_by
    right_shift = num >> shift_by
    return left_shift, right_shift

# Example test data
num = 8       # binary: 1000
shift_by = 2  # shift by 2 positions

ls, rs = shift_operations(num, shift_by)

print(f"Number: {num}")
print(f"Left Shift ({num} << {shift_by}) = {ls}")
print(f"Right Shift ({num} >> {shift_by}) = {rs}")
```

```python
# Program to check membership of a number in a sequence

def check_membership(num, sequence):
    if num in sequence:
        print(f"{num} is present in the sequence {sequence}")
    else:
        print(f"{num} is NOT present in the sequence {sequence}")

# Example test data
sequence = (10, 20, 56, 78, 89)
num = 56

check_membership(num, sequence)
```
