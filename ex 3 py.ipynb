```python
# Program to check whether a number is divisible by both 3 and 5

def check_divisibility(num):
    if num % 3 == 0 and num % 5 == 0:
        print(f"{num} is divisible by both 3 and 5")
    else:
        print(f"{num} is NOT divisible by both 3 and 5")

# Example test data
num = 30
check_divisibility(num)
```

```python
# Program to find the greatest among two numbers

def compare_numbers(a, b):
    if a > b:
        print(f"{a} is greater than {b}")
    elif b > a:
        print(f"{b} is greater than {a}")
    else:
        print("Numbers are equal")

# Example test data
a = 15
b = 20
compare_numbers(a, b)
```

```python
# Program to find the greatest among three numbers (all different)

def greatest_of_three(a, b, c):
    if a > b and a > c:
        print(f"{a} is the greatest")
    elif b > a and b > c:
        print(f"{b} is the greatest")
    else:
        print(f"{c} is the greatest")

# Example test data
a = 12
b = 25
c = 18
greatest_of_three(a, b, c)
```

```python
import math

# Program to check whether quadratic equation has real or imaginary roots

def quadratic_roots(a, b, c):
    D = b**2 - 4*a*c  # discriminant
    
    if D > 0:
        root1 = (-b + math.sqrt(D)) / (2*a)
        root2 = (-b - math.sqrt(D)) / (2*a)
        print("Roots are real and distinct:")
        print(f"Root1 = {root1}, Root2 = {root2}")
    elif D == 0:
        root = -b / (2*a)
        print("Roots are real and equal:")
        print(f"Root = {root}")
    else:
        real_part = -b / (2*a)
        imag_part = math.sqrt(-D) / (2*a)
        print("Roots are imaginary (complex):")
        print(f"Root1 = {real_part} + {imag_part}i")
        print(f"Root2 = {real_part} - {imag_part}i")

# Example test data
a = 1
b = 2
c = 5
quadratic_roots(a, b, c)
```

```python
# Program to check whether a given year is a leap year

def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        print(f"{year} is a leap year")
    else:
        print(f"{year} is NOT a leap year")

# Example test data
year = 2024
is_leap_year(year)
```

```python
from datetime import datetime, timedelta

# Program to display the next date in the calendar

def next_date(date_str):
    # Parse input date (format: DD-MM-YYYY)
    given_date = datetime.strptime(date_str, "%d-%m-%Y")
    
    # Add one day
    next_day = given_date + timedelta(days=1)
    
    # Display result
    print("Given Date:", given_date.strftime("%d-%m-%Y"))
    print("Next Date :", next_day.strftime("%d-%m-%Y"))

# Example test data
date_str = "28-02-2024"  # Leap year case
next_date(date_str)
```

```python
# Program to print grade sheet of a student from marks of five subjects

def grade_sheet(name, marks):
    # marks is a list of 5 subject marks
    total = sum(marks)
    average = total / len(marks)
    
    # Assign grade based on average
    if average >= 90:
        grade = "A+"
    elif average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 50:
        grade = "D"
    else:
        grade = "F"
    
    # Print grade sheet
    print("----------- Grade Sheet -----------")
    print(f"Student Name : {name}")
    for i, mark in enumerate(marks, start=1):
        print(f"Subject {i} Marks : {mark}")
    print(f"Total Marks  : {total}")
    print(f"Average Marks: {average:.2f}")
    print(f"Grade        : {grade}")
    print("-----------------------------------")

# Example test data
student_name = "Rahul"
marks = [85, 78, 92, 88, 76]  # marks in 5 subjects

grade_sheet(student_name, marks)
```
